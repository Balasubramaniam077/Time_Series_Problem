```python
#VIOLIN PLOT FOR TIME SERIES PROBLEM 



sns.set(style="darkgrid")
fig,axs=plt.subplots(2,2,figsize=(12,4))
sns.violinplot(data=df,x="close",kde=True,ax=axs[0,0],color='green')
sns.violinplot(data=df,x="high",kde=True,ax=axs[0,1],color='red')
sns.violinplot(data=df,x="low",kde=True,ax=axs[1,0],color='blue')
sns.violinplot(data=df,x="volume",kde=True,ax=axs[1,1],color='yellow')
```

```python
#HISTROGRAM PLOT
sns.set(style="darkgrid")
fig,axs=plt.subplots(2,2,figsize=(12,12))
sns.histplot(data=df,x="close",kde=True,ax=axs[0,0],color='green')
sns.histplot(data=df,x="high",kde=True,ax=axs[0,1],color='red')
sns.histplot(data=df,x="low",kde=True,ax=axs[1,0],color='blue')
sns.histplot(data=df,x="volume",kde=True,ax=axs[1,1],color='orange')
```

```python
#SKEWNESS USING BOXPLOT

plt.figure(figsize=(12,6))
sns.boxplot("close","low",data=df,palette="YlOrBr")
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
```
